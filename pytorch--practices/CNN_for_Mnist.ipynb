```python

import torch
from torch.autograd import Variable
import torch.nn as nn
import torch.nn.functional as F
import torchvision.datasets as dsets
import torchvision.transforms as transform

batch_size=64

train_dataset = dsets.MNIST(root='../data', train=True, transform=transform.ToTensor(), download=True)
train_loader = torch.utils.data.DataLoader(dataset=train_dataset, batch_size=batch_size, shuffle=False)
print(train_loader)

class Net(nn.Module):
    
    def __init__(self):
        super(Net,self).__init__()
        #kernel
        self.conv1 = nn.Conv2d(1,16,kernel_size=5,padding=2)
        self.norm1 = nn.BatchNorm2d(16)
        self.relu1 = nn.ReLU()
        self.mpool1 = nn.MaxPool2d(2)
        
        self.conv2 = nn.Conv2d(16,32,kernel_size=5,padding=2)
        self.norm2 = nn.BatchNorm2d(32)
        self.relu2 = nn.ReLU()
        self.mpool2 = nn.MaxPool2d(2)
        
        self.fc1 = nn.Linear(1568,10)
        
    def foward(self, x):
        x = self.conv1(x)
        x = self.norm1(x)
        x = self.relu1(x)
        x = self.mpool1(x)
        
        x = self.conv2(x)
        x = self.norm2(x)
        x = self.relu2(x)
        x = self.mpool2(x)
        
        x= x.view(64,-1)
        x = self.fc1(x)
        #print(x.size())
        return x
        
    def num_flat_features(self,x):
        pass
        


loss =nn.CrossEntropyLoss()
#criterion = nn.MSELoss()
n=Net()

#optimizer는 파라미터 업데이트를 자동으로해줌.
optimizer = torch.optim.Adam(n.parameters(), lr=1e-4)

output_loss = 0
for j in range(100):

    for i, (img,label) in enumerate(train_loader):
        #size 수정
        x=Variable(img, requires_grad=False)
        label=Variable(label, requires_grad=False)
        
        o = n.foward(x)
        output_loss = loss(o,label)
        n.zero_grad()
        output_loss.backward()
        optimizer.step()

        print(i,output_loss.data[0])
```
